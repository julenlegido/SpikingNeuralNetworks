```python
import json
with open("results/logs/ann_results_10.json") as f:
    ann = json.load(f)

with open("results/logs/snn_results_10.json") as f:
    snn = json.load(f)
```

```python
from matplotlib import pyplot as plt


plt.bar(["ANN", "SNN"], [ann["accuracy"][-1], snn["accuracy"][-1]])
plt.title("Accuracy Comparison")
plt.show()


plt.bar(["ANN", "SNN"], [
    sum(ann["time"]) / len(ann["time"]),
    sum(snn["time"]) / len(snn["time"])
])
plt.title("Time per Epoch")
plt.show()


plt.bar(["ANN", "SNN"], [0.00,0.00])
plt.title("Energy Consumption (CarbonTracker)")
plt.show()
```
