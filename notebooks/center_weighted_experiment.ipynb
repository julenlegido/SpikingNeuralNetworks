```python
import json
import matplotlib.pyplot as plt

# Standard encoding (scale = 0.5)
with open("results/logs/snn_rate_results_steps25_scale0.5.json") as f:
    standard = json.load(f)

with open("results/logs/spike_rate_results_steps25_scale0.5.json") as f:
    standard_spike = json.load(f)

# Center-weighted encoding
with open("results/logs/snn_center_results_scale0.5.json") as f:
    center = json.load(f)

with open("results/logs/spike_center_results_scale0.5.json") as f:
    center_spike = json.load(f)
```

```python
labels = ["Standard", "Center-weighted"]

accuracy = [
    standard["accuracy"][-1],
    center["accuracy"][-1]
]

total_spikes = [
    standard_spike["total_spikes"],
    center_spike["total_spikes"]
]

firing_rate = [
    standard_spike["firing_rate"],
    center_spike["firing_rate"]
]

sparsity = [
    standard_spike["sparsity"],
    center_spike["sparsity"]
]
```

```python
plt.figure()
plt.bar(labels, accuracy)

plt.ylabel("Accuracy (%)")
plt.title("Accuracy Comparison: Standard vs Center-weighted")

plt.savefig("results/figures/center_accuracy.png", dpi=300)
plt.show()
```

```python
plt.figure()
plt.bar(labels, total_spikes)

plt.ylabel("Total Spikes")
plt.title("Total Spikes Comparison")

plt.savefig("results/figures/center_total_spikes.png", dpi=300)
plt.show()
```

```python
plt.figure()
plt.bar(labels, firing_rate)

plt.ylabel("Firing Rate")
plt.title("Firing Rate Comparison")

plt.savefig("results/figures/center_firing_rate.png", dpi=300)
plt.show()
```

```python
plt.figure()
plt.bar(labels, sparsity)

plt.ylabel("Sparsity")
plt.title("Sparsity Comparison")

plt.savefig("results/figures/center_sparsity.png", dpi=300)
plt.show()
```

```python
import numpy as np

x = np.arange(len(labels))
width = 0.2

plt.figure(figsize=(8,5))

plt.bar(x, firing_rate, width, label="Firing Rate")
plt.bar(x + width, sparsity, width, label="Sparsity")

plt.xticks(x, labels)
plt.title("Comparison of Encoding Strategies")
plt.legend()

plt.savefig("results/figures/center_combined.png", dpi=300)
plt.show()
```
