```python
import json
import matplotlib.pyplot as plt

labels = ["Standard", "Center", "Gradient"]

# --- Standard ---
with open("results/logs/snn_rate_results_steps25_scale0.5.json") as f:
    standard = json.load(f)

with open("results/logs/spike_rate_results_steps25_scale0.5.json") as f:
    standard_spike = json.load(f)

# --- Center ---
with open("results/logs/snn_center_results_steps25_scale0.5.json") as f:
    center = json.load(f)

with open("results/logs/spike_center_results_steps25_scale0.5.json") as f:
    center_spike = json.load(f)

# --- Gradient ---
with open("results/logs/snn_gradient_results_prob0.5.json") as f:
    gradient = json.load(f)

with open("results/logs/spike_gradient_results_scale0.5.json") as f:
    gradient_spike = json.load(f)
```

```python
accuracy = [
    standard["accuracy"][-1],
    center["accuracy"][-1],
    gradient["accuracy"][-1]
]

total_spikes = [
    standard_spike["total_spikes"],
    center_spike["total_spikes"],
    gradient_spike["total_spikes"]
]

firing_rate = [
    standard_spike["firing_rate"],
    center_spike["firing_rate"],
    gradient_spike["firing_rate"]
]

sparsity = [
    standard_spike["sparsity"],
    center_spike["sparsity"],
    gradient_spike["sparsity"]
]
```

```python
plt.figure()
plt.bar(labels, accuracy)

plt.ylabel("Accuracy (%)")
plt.title("Accuracy Comparison")

plt.savefig("results/figures/gradient_accuracy.png", dpi=300)
plt.show()
```

```python
plt.figure()
plt.bar(labels, total_spikes)

plt.ylabel("Total Spikes")
plt.title("Total Spikes Comparison")

plt.savefig("results/figures/gradient_total_spikes.png", dpi=300)
plt.show()
```

```python
plt.figure()
plt.bar(labels, firing_rate)

plt.ylabel("Firing Rate")
plt.title("Firing Rate Comparison")

plt.savefig("results/figures/gradient_firing_rate.png", dpi=300)
plt.show()
```

```python
plt.figure()
plt.bar(labels, sparsity)

plt.ylabel("Sparsity")
plt.title("Sparsity Comparison")

plt.savefig("results/figures/gradient_sparsity.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(labels, firing_rate, marker='o', label="Firing Rate")
plt.plot(labels, sparsity, marker='o', label="Sparsity")

plt.title("Encoding Comparison (MNIST)")
plt.legend()
plt.grid()

plt.savefig("results/figures/gradient_combined.png", dpi=300)
plt.show()
```
