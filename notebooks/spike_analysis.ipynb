```python
import json

timesteps = [10, 25, 50, 100]
firing_rates = []
sparsities = []
total_spikes = []

for t in timesteps:
    with open(f"results/logs/spike_results_{t}.json") as f:
        data = json.load(f)

    firing_rates.append(data["firing_rate"])
    sparsities.append(data["sparsity"])
    total_spikes.append(data["total_spikes"])
```

```python
import matplotlib.pyplot as plt

plt.plot(timesteps, firing_rates, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Firing Rate")
plt.title("Firing Rate vs Timesteps")
plt.grid()
plt.show()

plt.plot(timesteps, sparsities, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Sparsity")
plt.title("Sparsity vs Timesteps")
plt.grid()
plt.show()

plt.plot(timesteps, total_spikes, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Total Spikes")
plt.title("Total Spikes vs Timesteps")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(timesteps, firing_rates, marker='o', label="Firing Rate")
plt.plot(timesteps, sparsities, marker='o', label="Sparsity")

plt.xlabel("Timesteps")
plt.ylabel("Value")
plt.title("Spike Activity vs Timesteps")
plt.legend()
plt.grid()
plt.show()
```
