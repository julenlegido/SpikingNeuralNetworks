```python
import json

timesteps = [10, 25, 50, 100]
accuracies = []
times = []

for t in timesteps:
    with open(f"results/logs/snn_results_{t}.json") as f:
        data = json.load(f)

    accuracies.append(data["accuracy"][-1])
    times.append(sum(data["time"]) / len(data["time"]))
```

```python
import matplotlib.pyplot as plt

plt.plot(timesteps, accuracies, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy vs Timesteps")
plt.grid()
plt.show()

plt.plot(timesteps, times, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Time per Epoch (s)")
plt.title("Time vs Timesteps")
plt.grid()
plt.show()
```
