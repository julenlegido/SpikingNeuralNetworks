```python
import json

timesteps = [10, 25, 50, 100]
accuraciesSNN = []
timesSNN = []

for t in timesteps:
    with open(f"results/logs/snn_results_{t}.json") as f:
        data = json.load(f)

    accuraciesSNN.append(data["accuracy"][-1])
    timesSNN.append(sum(data["time"]) / len(data["time"]))
```

```python
import json

timesteps = [10, 25, 50, 100]
accuraciesANN = []
timesANN = []

for t in timesteps:
    with open(f"results/logs/ann_results_{t}.json") as f:
        data = json.load(f)

    accuraciesANN.append(data["accuracy"][-1])
    timesANN.append(sum(data["time"]) / len(data["time"]))
```

```python
import matplotlib.pyplot as plt

plt.plot(timesteps, accuraciesSNN, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Accuracy SNN (%)")
plt.title("Accuracy SNN vs Timesteps")
plt.grid()
plt.show()

plt.plot(timesteps, accuraciesANN, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("Accuracy ANN (%)")
plt.title("Accuracy ANN vs Timesteps")
plt.grid()
plt.show()

plt.plot(timesteps, timesSNN, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("SNN Time per Epoch (s)")
plt.title("SNN Time vs Timesteps")
plt.grid()
plt.show()

plt.plot(timesteps, timesANN, marker='o')
plt.xlabel("Timesteps")
plt.ylabel("ANN Time per Epoch (s)")
plt.title("ANN Time vs Timesteps")
plt.grid()
plt.show()
```
