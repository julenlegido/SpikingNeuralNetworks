```python
import json
with open("results/logs/snn_results_10.json") as f:
    snn = json.load(f)

with open("results/logs/ann_results_10.json") as f:
    ann = json.load(f)
```

```python
from matplotlib import pyplot as plt


plt.plot(snn["loss"], label="SNN", marker='o')
plt.plot(ann["loss"], label="ANN", marker='o')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss per Epoch")
plt.legend()
plt.grid()
plt.show()
```
